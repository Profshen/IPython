# 加载数据

#### 从本地目录加载移动通话数据，共计2303140条通话记录（source＝ 主叫， target＝ 被叫）

```python
edge_data = gl.SFrame.read_csv("./downloads/mygraphlab/yichangjituan.csv",column_type_hints=[str,str,str,float])
```

### 数据预处理：命名字段、抽取日期、求和、求平均、汇总

```python
edge_data.rename({'date':'timestamp'})  # 将日期变量重新命名为时间戳 timestamp  ，不是一定要这样
```

```python
# 显示边数据情况，发现共有59024个电话号码呼叫，60203个被叫电话号码，通话时长最小是1秒，最长通话时长是21660
edge_data.show()
```

### 从时间戳字段抽取：年、月、日

```python
edge_data['timestamp']=edge_data['timestamp'].str_to_datetime('%Y-%m-%d-%H-%M-%S')
```

```python
edge_data.add_columns(edge_data['timestamp'].split_datetime(column_name_prefix=None,limit=['year','month','day']))
```

### 根据source，target字段进行汇总，计算任意两个电话号码的通话次数

```python
# 汇总计算，将source和target汇总（月通话次数）
sf = edge_data.groupby(['source','target'],{'calltime_count' : gl.aggregate.COUNT, 'calltime_sum' : gl.aggregate.SUM('calltime')})
```

```python
# 查看汇总和的数据结构，共计一个月内有33万多不重复通话边记录
sf
```

```python
sf.sort('calltime_sum', ascending=False) # 按照月通话时长汇总排序
```

### 根据通话总时长和通话次数，计算平均每次通话时长

```python
sf['calltime_ave']=sf.apply(lambda x: x['calltime_sum']/x['calltime_count']) # 计算平均通话时长
```

```python
sf.sort('calltime_sum', ascending=False) # 按照月平均通话时长计算排序
```

```python
sf.sort('calltime_count', ascending=False) # 按照月通话次数排序
```

```python
sf.sort('calltime_ave', ascending=False) # 某两个电话号码一个月内通话只有一次，但通话时长达到21660最大值
```

## # 采用networkx展示layout

```python
import networkx as nx
ng = nx.Graph()

# Put the nodes and edges from the SGraph into a NetworkX graph
ng.add_nodes_from(list(kol_net1.vertices['__id']))
ng.add_edges_from([(e['__src_id'], e['__dst_id']) for e in kol_net1.edges])

# Create the layout with NetworkX and convert to regular Python types
# You can substitute any of the layout algorithms here for circular_layout:
# http://networkx.github.io/documentation/latest/reference/drawing.html#module-networkx.drawing.layout
layout = nx.circular_layout(ng)
layout = {k: map(float, list(v)) for k,v in layout.iteritems()}

# Show the SGraph in Canvas with that layout
kol_net1.vertices['x'] = kol_net1.vertices.apply(lambda v: layout[v['__id']][0])
kol_net1.vertices['y'] = kol_net1.vertices.apply(lambda v: layout[v['__id']][1])
kol_net1.show(vertex_positions=('x', 'y'),vlabel='id', highlight=[kol_1], elabel='calltime_count',
          arrows=True, vlabel_hover=False, elabel_hover=False)
```

```python
import time

def pagerank_update_fn(src, edge, dst):
    dst['pagerank'] += src['prev_pagerank'] * edge['weight']
    return (src, edge, dst)

def sum_weight(src, edge, dst):
    src['total_weight'] += edge['weight']
    return src, edge, dst

def normalize_weight(src, edge, dst):
    edge['weight'] /= src['total_weight']
    return src, edge, dst

def pagerank_triple_apply(input_graph, reset_prob=0.15, threshold=1e-3, 
                          max_iterations=20):
    g = gl.SGraph(input_graph.vertices, input_graph.edges)

    # compute normalized edge weight
    g.vertices['total_weight'] = 0.0
    g = g.triple_apply(sum_weight, ['total_weight'])
    g = g.triple_apply(normalize_weight, ['weight'])
    del g.vertices['total_weight']

    # initialize vertex field
    g.vertices['prev_pagerank'] = 1.0
    it = 0
    total_l1_delta = len(g.vertices)
    start = time.time()
    while(total_l1_delta > threshold and it < max_iterations):
        g.vertices['pagerank'] = 0.0
        g = g.triple_apply(pagerank_update_fn, ['pagerank'])
        g.vertices['pagerank'] = g.vertices['pagerank'] * (1 - reset_prob) \
                                                                  + reset_prob
        g.vertices['l1_delta'] = (g.vertices['pagerank'] - \
                          g.vertices['prev_pagerank']).apply(lambda x: abs(x))
        total_l1_delta = g.vertices['l1_delta'].sum()
        g.vertices['prev_pagerank'] = g.vertices['pagerank']
        print 'Iteration %d: total pagerank changed in L1 = %f' % (it,\
                                                                total_l1_delta)
        it = it + 1
    print 'Triple apply pagerank finished in: %f secs' % (time.time() - start)
    del g.vertices['prev_pagerank']
    return g

# Load graph
# g = gl.load_graph('http://snap.stanford.edu/data/email-Enron.txt.gz', 'snap')
# g.edges['weight'] = 1.0
```
