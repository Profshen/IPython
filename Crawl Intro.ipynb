urllib 和urllib2都是接受URL请求的相关模块，但是提供了不同的功能.</br>
>urllib2可以接受一个Request对象，并以此可以来设置一个URL的headers，但是urllib只接收一个URL。这意味着，你不能伪装你的用户代理字符串等。</br>
>urllib模块可以提供进行urlencode的方法，该方法用于GET查询字符串的生成，urllib2的不具有这样的功能。这就是urllib与urllib2经常在一起使用的原因。

```python
import urllib
import urllib2
```

urllib2.Request(url[, data][, headers][, origin_req_host][, unverifiable])</br>
Request类是一个抽象的URL请求。有5个参数：</br>
>URL——是一个字符串，其中包含一个有效的URL。</br>
>data——是一个字符串，指定额外的数据发送到服务器，如果没有data需要发送可以为“None。数据需要被以标准的格式编码（encode），然后作为一个数据参数传送给Request对象。Encoding是在urlib模块中完成的。当请求含有data参数时，HTTP的请求为POST，而不是GET。</br>
>headers——是字典类型，头字典可以作为参数在request时直接传入,因为一些HTTP服务只允许某些请求来自常见的浏览器而不是脚本，所以需要用headers来伪装辨别浏览器身份的User-Agent。

# 通过API抓取数据

## 1、Get 方法

GET - 从指定的资源请求数据。查询字符串（名称/值对）是在 GET 请求的 URL 中发送的


```python
url = "https://api.weibo.com/2/statuses/repost_timeline.json"
headers = { 'User-Agent' : 'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)' }
params = {"id":3787234660856097,
          "count":200,
        "access_token":"2.00fDbgnBJUbr3Efa77e4013dykHqOC"}
data = urllib.urlencode(params)
print data
urls = url + "?"+data
print "urls:",urls
```

```python
req = urllib2.Request(urls,None,headers)
response = urllib2.urlopen(req)
result = response.read()
print result
```

## 2、Post方法

POST - 向指定的资源提交要被处理的数据。查询字符串（名称/值对）是在 POST 请求的 HTTP 消息主体中发送的

```python
url = "https://api.weibo.com/2/comments/create.json"
headers = { 'User-Agent' : 'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)' }
params = {"id":3787120281391909,
        "access_token":"2.00fDbgnBJUbr3Efa77e4013dykHqOC",
        "comment":"数据抓取post方法，by小和子"}
data = urllib.urlencode(params)
print data
```

```python
req = urllib2.Request(url,data,headers)
response = urllib2.urlopen(req)
result = response.read()
print result
```

## 3、Json数据解析

JSON(JavaScript Object Notation) 是一种轻量级的数据交换格式。易于人阅读和编写。同时也易于机器解析和生成。</br>
JSON结构：“名称/值”对的集合（A collection of name/value pairs）</br>
>例：{'name' : 'terry tan' , 'school' : 'CUC'}</br>
json.dumps方法对简单数据类型进行编码,即将json解码成string</br>
json.loads方法对简单数据类型进行解码，即将string解码成json

```python
import json
```

```python
j = json.loads(result)
total = j["total_number"]
out_file = open("c:/users/tanhe/desktop/weibo.txt","w")
for i in range(0,total):
    usrname = j["reposts"][i]["user"]["name"]
    text = j["reposts"][i]["text"]
    repost = j["reposts"][i]["reposts_count"]
    row = usrname + "\t"+ text +"\t"+ str(repost) + "\n"
    out_file.write(row)
out_file.close()
```
